A palindrome number is a number that remains the same when its digits are reversed. For example, 121 is a palindrome, but 123 is not. The following code will take an integer input and check if it's a palindrome.

```python
def is_palindrome(n):
    # Convert the number to a string to easily reverse it
    s = str(n)
    # Compare the string with its reverse
    return s == s[::-1]

# Get input from the user
while True:
    try:
        num = int(input("Enter an integer: "))
        break
    except ValueError:
        print("Invalid input. Please enter an integer.")

# Check if the number is a palindrome and print the result
if is_palindrome(num):
    print(f"{num} is a palindrome number.")
else:
    print(f"{num} is not a palindrome number.")
```
