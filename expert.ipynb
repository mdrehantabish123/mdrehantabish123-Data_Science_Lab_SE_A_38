## Implement Inference Engine

### Subtask:
Write Python code to implement a basic inference engine that can process the facts and apply the defined rules to reach a decision.


**Reasoning**:
Implement a basic inference engine by defining a list of rules and a function to apply these rules to given facts to recommend a drink. This covers all instructions in the subtask.



```python
rules = [
    {
        "conditions": {
            "weather_condition": "hot",
            "user_preference_type": "caffeinated"
        },
        "consequence": "Iced Coffee"
    },
    {
        "conditions": {
            "weather_condition": "hot",
            "user_preference_type": "decaffeinated"
        },
        "consequence": "Iced Tea"
    },
    {
        "conditions": {
            "weather_condition": "cold",
            "user_preference_type": "caffeinated"
        },
        "consequence": "Hot Coffee"
    },
    {
        "conditions": {
            "weather_condition": "cold",
            "user_preference_type": "decaffeinated"
        },
        "consequence": "Hot Chocolate"
    },
    {
        "conditions": {
            "weather_condition": "mild",
            "user_preference_type": "caffeinated"
        },
        "consequence": "Latte"
    },
    {
        "conditions": {
            "weather_condition": "mild",
            "user_preference_type": "decaffeinated"
        },
        "consequence": "Herbal Tea"
    }
]

def recommend_drink(facts):
    """
    Recommends a drink based on given facts by applying a set of rules.

    Args:
        facts (dict): A dictionary containing 'weather_condition' and 'user_preference_type'.

    Returns:
        str: The recommended drink or a default message if no rule matches.
    """
    for rule in rules:
        conditions_met = True
        for condition_key, condition_value in rule["conditions"].items():
            if facts.get(condition_key) != condition_value:
                conditions_met = False
                break
        if conditions_met:
            return rule["consequence"]
    return "No specific recommendation available based on the given facts."

print("Inference engine rules and function 'recommend_drink' defined.")
```
