```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

# ===============================
# LOAD DATASETS
# ===============================
print("\n===== Loading Datasets =====\n")

daily_df = pd.read_csv('worldometer_coronavirus_daily_data.csv.zip')

print("Worldometer Coronavirus Daily Data")
display(daily_df.head())

# ===============================
# DAILY DATA OVERVIEW
# ===============================
print("\n===== Daily Data Overview =====\n")

print("\n--- daily_df info() ---\n")
daily_df.info()

print("\n--- daily_df describe() ---\n")
display(daily_df.describe())

# ===============================
# DATA CLEANING
# ===============================
print("\n===== Data Cleaning =====\n")

daily_df['date'] = pd.to_datetime(daily_df['date'])
daily_df.fillna(0, inplace=True)

daily_df['daily_new_cases'] = daily_df['daily_new_cases'].apply(lambda x: max(0, x))
daily_df['daily_new_deaths'] = daily_df['daily_new_deaths'].apply(lambda x: max(0, x))
daily_df['active_cases'] = daily_df['active_cases'].apply(lambda x: max(0, x))

# ===============================
# AFTER CLEANING OUTPUT
# ===============================
print("\n===== After Cleaning Output =====\n")

print("\n--- daily_df info() after cleaning ---\n")
daily_df.info()

print("\n--- daily_df describe() after cleaning ---\n")
display(daily_df.describe())

# ===============================
# EXTRA CHECKS
# ===============================
print("\n===== Additional Checks =====\n")

print("\nMissing Values:\n")
print(daily_df.isnull().sum())

print("\nTotal Countries:", daily_df['country'].nunique())

print("\n✅ All Outputs Generated Successfully!")
```
