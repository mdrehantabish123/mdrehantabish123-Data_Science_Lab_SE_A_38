First, we'll ask the user to input a number. Then, we'll check if the number is even or odd using the modulo operator (`%`). If a number divided by 2 has a remainder of 0, it's even; otherwise, it's odd.

```python
# Get input from the user
num = int(input("Enter a number: "))

# Check if the number is even or odd
if (num % 2) == 0:
   print(f"{num} is an even number")
else:
   print(f"{num} is an odd number")
```
