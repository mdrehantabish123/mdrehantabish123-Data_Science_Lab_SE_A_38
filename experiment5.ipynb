```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Map numerical species to names for better readability
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_map)

print("1. First 5 rows of the Iris DataFrame:\n")
# Display the first 5 rows of the iris_df DataFrame
print(iris_df.head())

print("\n2. Concise summary of the DataFrame (info()):\n")
# Display concise summary of the DataFrame
iris_df.info()

print("\n3. Descriptive statistics of numerical columns (describe()):\n")
# Generate descriptive statistics of the numerical columns
print(iris_df.describe())
```

```python
print("\n4. Filtering data where 'petal length (cm)' > 4.0:\n")
# Filter the iris_df DataFrame to select rows where 'petal length (cm)' is greater than 4.0
filtered_iris_df = iris_df[iris_df['petal length (cm)'] > 4.0]

# Display the first 5 rows of the filtered_iris_df DataFrame
print(filtered_iris_df.head())

print("\n5. Grouping by species and computing mean of numerical columns:\n")
# Group the original iris_df DataFrame by the 'species' column and compute the mean
grouped_mean = iris_df.groupby('species').mean(numeric_only=True)

# Display the grouped_mean DataFrame
print(grouped_mean)

print("\n6. Grouping by species and computing sum, count, min, and max for numerical columns:\n")
# Group by species and compute sum, count, min, and max for numerical columns
aggregated_stats = iris_df.groupby('species').agg(['sum', 'count', 'min', 'max'])
print(aggregated_stats)
```
